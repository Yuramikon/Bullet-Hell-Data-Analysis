### <span style="color: #2196F3;">Real-World Problem & Significance</span>
The **"Bullet Hell"** genre, which originated in Japan's arcade culture, has become increasingly difficult for players to discover because its Steam tag has been "hijacked" by unrelated game genres. Since **Steam** plays a vital role in the continued survival and accessibility of this genre today, distinguishing between games that remain true to their roots and those that do not is essential for both preservation and user experience.

### <span style="color: #4CAF50;">Prediction Target & Task Type</span>
In this project, I will attempt to predict a feature named <span style="color: #d9534f; font-weight: bold;">"real"</span> to differentiate authentic Bullet Hell titles within this Steam tag. 

* **Task Type:** This is a <span style="color: #8e44ad; font-weight: bold;">Classification Task</span>.
* **Reasoning:** The target variable is categorical and binary, consisting of two distinct classes: 
    * **1:** Authentic Bullet Hell titles.
    * **0:** Unrelated or "mis-tagged" titles.

</div>

## 3. Data Loading & Initial Inspection


```python
import pandas as pd

df = pd.read_csv("ShmupData.csv")

print("-" * 20, "First 5 Rows", "-" * 20)
display(df.head())

print("\n" + "-" * 20, "Dataset Information", "-" * 20)
print(f"Dataset Shape: {df.shape[0]} rows and {df.shape[1]} columns")
display(df.info())

print("\n" + "-" * 20, "Statistical Summary", "-" * 20)
display(df.describe())
```

## 4. Data Cleaning

```python
# --- STEP 1: Handling Missing Values ---
# Justification: The 'real' feature is our target. Rows without a target value provide no learning signal.
initial_count = len(df)
df = df.dropna(subset=['real'])
print(f"1. Missing Values: Dropped {initial_count - len(df)} rows with missing 'real' values.")
```

```python
# --- STEP 2: Duplicate Rows ---
duplicates = df.duplicated().sum()
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"2. Duplicates: Removed {duplicates} duplicate rows.")
else:
    print("2. Duplicates: No duplicate rows found.")
```

```python
# --- STEP 3: Boolean to Integer Conversion ---
# Justification: Machine learning models require numerical input. 
bool_cols = ['win', 'mac', 'linux', 'steam_deck']
for col in bool_cols:
    if col in df.columns:
        df[col] = df[col].astype(int)
print(f"3. Dtypes: Converted {bool_cols} from boolean to integer.")
```

```python
# --- STEP 4: Inconsistent Categories & Formatting ---
# Justification: Ensuring text consistency (stripping spaces, standardizing case).
if 'rating' in df.columns:
    df['rating'] = df['rating'].str.strip().str.title()
print("4. Categories: Standardized 'rating' labels.")
```

```python
# --- STEP 5: Reporting the Most Extreme Outliers ---
# Justification: Highlighting the single most distant data point from IQR bounds 
# to identify the most significant anomalies in the dataset.

print("5. Outlier Detection")

def report_most_extreme(data, column, title_col='title'):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Calculate distance from bounds
    upper_outliers = data[data[column] > upper_bound]
    lower_outliers = data[data[column] < lower_bound]
    
    print(f"\n" + "="*60)
    print(f"EXTREME ANOMALY REPORT: {column.upper()}")
    print(f"Standard Bounds: {lower_bound:.2f} to {upper_bound:.2f}")
    
    # Find the single most extreme upper outlier
    if not upper_outliers.empty:
        extreme_upper = upper_outliers.loc[upper_outliers[column].idxmax()]
        distance = extreme_upper[column] - upper_bound
        print(f"Most Extreme (High): '{extreme_upper[title_col]}'")
        print(f"Value: {extreme_upper[column]} | Distance from Bound: +{distance:.2f}")
    
    # Find the single most extreme lower outlier
    if not lower_outliers.empty:
        extreme_lower = lower_outliers.loc[lower_outliers[column].idxmin()]
        distance = lower_bound - extreme_lower[column]
        print(f"Most Extreme (Low): '{extreme_lower[title_col]}'")
        print(f"Value: {extreme_lower[column]} | Distance from Bound: +{distance:.2f}")
        
    if upper_outliers.empty and lower_outliers.empty:
        print("No outliers detected for this feature.")
    print("="*60)

# Execute report for key metrics
for feature in ['user_reviews', 'price_original', 'positive_ratio']:
    if feature in df.columns:
        report_most_extreme(df, feature)
```

After inspecting the outliers. I will keep these for now as they represent authentic (though extreme) market performance.

### <span style="color: #F44336;">5. Univariate Analysis: Feature Dropping</span>

During the univariate analysis, it was observed that the `win` and `steam_deck` features contain the same value (1) for all records in the filtered dataset. 

* **Justification:** Columns with zero variance do not provide any predictive power or information gain for a classification model. Removing these redundant features simplifies the dataset and improves computational efficiency without losing any meaningful information.

```python
cols_to_drop = ['win', 'steam_deck']

existing_cols_to_drop = [col for col in cols_to_drop if col in df.columns]
df = df.drop(columns=existing_cols_to_drop)
```

```python
# --- 4. TAGS ANALYSİS ---
import ast

df_tags = df.copy()
df_tags['tags'] = df_tags['tags'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

all_tags = df_tags.explode('tags')

tags_to_exclude = ["Shoot 'Em Up", "Bullet Hell"]
filtered_tags = all_tags[~all_tags['tags'].isin(tags_to_exclude)]

top_distinctive_tags = filtered_tags['tags'].value_counts().head(20)

plt.figure(figsize=(12, 7))
sns.barplot(
    x=top_distinctive_tags.values, 
    y=top_distinctive_tags.index, 
    hue=top_distinctive_tags.index, 
    palette='viridis', 
    legend=False
)

plt.title('Top 20 Most Frequent Distinctive Tags (Excluding Core Genre Tags)')
plt.xlabel('Number of Games')
plt.ylabel('Tag Name')
plt.show()
```

```python
# --- 3. Multivariate Tag Analysis ---
# Analyzing which tags are most associated with "Real" Bullet Hell games
df_tags = df.copy()
df_tags['tags'] = df_tags['tags'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
exploded_tags = df_tags.explode('tags')

# Filter out the core genre tags again for a cleaner analysis
exclude = ["Shoot 'Em Up", "Bullet Hell"]
tag_impact = exploded_tags[~exploded_tags['tags'].isin(exclude)]

# Calculate the mean of 'real' for each tag (Percentage of being 'real' bullet hell)
tag_real_relation = tag_impact.groupby('tags')['real'].agg(['mean', 'count'])
# Filter for tags that appear in at least 20 games to ensure statistical significance
significant_tags = tag_real_relation[tag_real_relation['count'] >= 20].sort_values(by='mean', ascending=False)

# Visualize Top 20 Tags Most Correlated with "Real" status
plt.figure(figsize=(12, 7))
sns.barplot(x=significant_tags['mean'].head(20), y=significant_tags.head(20).index, palette='magma', hue=significant_tags.head(20).index, legend=False)
plt.title('Top 20 Tags with Highest Correlation to "Real" Bullet Hell Status')
plt.xlabel('Probability of being "Real"')
plt.show()

print("Statistical Summary of Top Tags:")
print(significant_tags.head(10))
```

```python
df_tags = df.copy()
df_tags['tags'] = df_tags['tags'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
exploded = df_tags.explode('tags')

exclude_tags = ["Shoot 'Em Up", "Bullet Hell"]
tag_stats = exploded[~exploded['tags'].isin(exclude_tags)].groupby('tags')['real'].agg(['mean', 'count'])

least_correlated_tags = tag_stats[tag_stats['count'] >= 20].sort_values(by='mean', ascending=True)

plt.figure(figsize=(12, 7))
sns.barplot(
    x=least_correlated_tags['mean'].head(20), 
    y=least_correlated_tags.head(20).index, 
    palette='flare', 
    hue=least_correlated_tags.head(20).index, 
    legend=False
)

plt.title('Top 20 Tags with LOWEST Correlation to "Real" Bullet Hell Status')
plt.xlabel('Probability of being "Real" (0.0 = Likely Unrelated)')
plt.ylabel('Tag Name')
plt.show()

print("Statistical Summary of Least Correlated Tags (Negative Predictors):")
print(least_correlated_tags.head(10))
```

## 7. Key Findings & Insights

* **The Core Niche:** Games that stay true to their **Bullet Hell** roots are fewer in number and tend to have a smaller volume of user reviews. However, they enjoy a higher appreciation rate (positive ratio) and are slightly more expensive than average titles.
* **Feature Importance:** It appears that **tags** will provide the most significant predictive power for our analysis.
* **Counter-intuitive Findings:** The low correlation of the **"Perma Death"** tag with the "real" status was unexpected, as it is traditionally associated with the genre's difficulty.
* **Hardware Standardization:** Univariate analysis revealed that **100%** of the titles in the dataset support both **Steam Deck** and **Windows**, making them universal constants rather than distinguishing features.
